```python
#import important libraries
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("/Users/Dell/Desktop/WEMC WORK/Book2 rainy.csv")
df.head()
df.drop("#",axis=1,inplace=True)
df.groupby("time").mean()
```

```python
df.head()
```

```python
df.groupby("time").mean().to_csv('/Users/Dell/Documents/new.csv')
```

```python
df.plot()
```

```python
df.plot(kind="line")
```

```python
plt.figure(figsize=(12,8))
sns.lineplot(data=df.groupby("time").mean(),x="time",y="value",color="orange",markers=True)
plt.title("Hourly variation of PM 2.5 in Rainy Season")
plt.show()
```
